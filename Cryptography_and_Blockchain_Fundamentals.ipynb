```python
import hashlib
from cryptography.hazmat.primitives import hashes, serialization
from cryptography.hazmat.primitives.asymmetric import rsa, padding


vehicles = {}


private_key = rsa.generate_private_key(public_exponent=65537, key_size=2048)
public_key = private_key.public_key()

# SHA 256 Function Definition
def genarte_sha256(data):
    sha256 = hashlib.sha256(data.encode()).hexdigest()
    return sha256


while True:
    print("\nbutton 1 for generate SHA-256 Hash")
    print("button 2 for Create & Verify Digital Signature")
    print("button 3 for Register a Vehicle")
    print("button 4 for Retrieve Vehicle Details")
    button = input("Enter option:")
    button = int(button)
    if button == 1:
        print("Generate SHA-256 Hash\n")
        data = input("Press the data:")
        sha = genarte_sha256(data)
        print(f" the sha vlaue is {data}  is : {sha}")
    elif button == 2:
        print("Create & Verify Digital Signature \n")
        Sign = input('Entre te data:')
        message = Sign.encode()

        # Sign the data with private key
        signature = private_key.sign(
            message,
            padding.PSS(
                mgf=padding.MGF1(hashes.SHA256()),
                salt_length=padding.PSS.MAX_LENGTH
            ),
            hashes.SHA256()
        )
        print(f"Signature (Hex): {signature.hex()}")

        # Verify the signature using public key
        try:
            public_key.verify(
                signature,
                message,
                padding.PSS(
                    mgf=padding.MGF1(hashes.SHA256()),
                    salt_length=padding.PSS.MAX_LENGTH),hashes.SHA256())
            print("Signature is valid")
        except:
            print("Signature is invalid")

    elif (button == 3):
        print(" Register a Vehicle")
        plate = input("Enter vehicle number plate:")


        if plate in vehicles:
            print(" Error: This number plate is already registered!")
        else:
            name = input("Enter owner's name: ")
            model = input("Enter vehicle model: ")


            vehicles[plate] = {'owner': name, 'model': model}
            print(f" Vehicle {plate} successfully registered.")

    elif button == 4:
        print("\Retrieve Vehicle Details ")
        plate = input("Enter number plate to search:").strip()

        # Look up the plate dictionary
        if plate in vehicles:
            details = vehicles[plate]
            print("Vehicle Details Found")
            print(f"Plate:{plate}")
            print(f"Owner:{details['owner']}")
            print(f"Model:{details['model']}")
        else:
            print(" Error: Vehicle number plate not found.")

    elif(button != 1 and button != 2 and button != 3 and button != 4):
      print("#########################  Entre a valid number ###############################3")
      break


```
